```python
#Reading in csv file first 5 rows

# Dont Hardcode the dataset, do this as the line below!
dataset = r'data.csv'
df = pd.read_csv(dataset)
#print out first few lines
print(df)
```

```python
#printing first 10 rows of dataset
printNumRows = pd.read_csv(dataset, nrows=10)
print(printNumRows)
```

```python
#Printing the whole dataset to get the mean/mdedian/mode of the column "new_cases"
#Using df short for dataframe of the CSV file

# https://datatofish.com/use-pandas-to-calculate-stats-from-an-imported-csv-file/ - assistance with the below code.

# The arithmetic mean is found by adding the numbers and dividing the sum by the number of numbers in the list. This is what is most often meant by an average. 
# The median is the middle value in a list ordered from smallest to largest. 
# The mode is the most frequently occurring value on the list. - https://www.britannica.com/science/mean-median-and-mode

avg = df['new_cases'].mean()
med = df['new_cases'].median()
mode = df['new_cases'].mode()
print("Average is : ",avg)
print()
print("Median is : ",med)
print()
print("Mode is : ",mode)
```

```python
# https://www.geeksforgeeks.org/how-to-plot-bar-graph-in-python-using-csv-file/ - BarPlots
# https://stackoverflow.com/questions/30482071/how-to-calculate-mean-values-grouped-on-another-column-in-pandas - group by
# https://www.google.com/search?q=iloc+in+python+means&rlz=1C1CHBF_enIE924IE924&oq=iloc+in+python&aqs=chrome.1.69i57j0i512l5j0i22i30l4.6078j0j7&sourceid=chrome&ie=UTF-8 - what is iloc?

# Trying to filter the new_cases column of all country codes (Averages are correct as per the csv)
# as index  = false gives the professional format as shown below
df.groupby('country_code', as_index=False)['new_cases'].mean()
```

```python
#Poor bar plot showing all the country codes with the total number of new cases they have.
# iloc enables you to pick a cell from the dataset you want to use i.e. country codes.
X = list(df.iloc[:, 1])
Y = list(df.iloc[:, 4])

plt.bar(X, Y, color='g')
plt.xlabel("new cases")
plt.ylabel("Country Code")
plt.title("new cases in countries around europe")
plt.show()
```
